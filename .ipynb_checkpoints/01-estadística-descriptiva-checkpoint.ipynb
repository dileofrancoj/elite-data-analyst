# Estadística descriptiva

## Conceptos base

### Población

Es el conjunto total de objetos que tienen características específicas que se quieren estudiar

### Muestra

Es un subconjunto de esa población que se toma como fuente de estudio

## Areas de la estadística

- Descriptiva: Se encarga de describir, visualizar y resumir datos que son originados a partir de diferentes fenómenos. El objetivo de esta rama es resumir las características principales de un conjunto de datos



- Inferencial: Utiliza y genera modelos matemáticos y predicciones asociadas a fenomenos teniendo en cuenta la aleatoriedad de las observaciones. A partir de las características de una muestra, infiere como se va a comportar una población más grande.

## Tipos de datos

- Cualitativo

Son propiedades categóricas que pueden usarse para describir un ente (nacionalidad, tipo de auto, etc)

- Cuantitativo

Representan una cantidad. Ej: presión arteríal, pulsaciones, distancia, etc.

### Tipos de variables

- Cualitativa nominal: Los valores que toma esa variable aleatoria permite clasificar a la población en categorías excluyentes. Grupo sanguineo, estado civil, etc.

- Cualitativa ordinal: Los valores que toma la VA pueden ser ordenados en base a un criterio. XEJ En peya distinguimos segmentos de usuario en base a la actividad y a la frecuencia de compra. Estables, nuevos, ocasionales. Estos tienen un orden intrinseco

- Cuantitativa discreta: Pueden tomar valores numéricos en un intervalo Natural, entero. Cantidad de compras que tiene un usuario, cantidad de usuarios por minuto que entra a la aplicacion, etc.

- Cuantitativa continua: Pueden tomar todos los valores de un intervalo real. Tiempo que tardó la aplicación en cargar, distancia entre el lugar y el usuario, mediciones de tiempo, etc.

### Outlier o Valor atípico

En estadística, un valor atípico es una observación que numéricamente es distante al resto de los datos.

Imaginemos el caso del barrio de Medina, y queremos calcular la media de la variable aleatoria salario. Acá tenemos un gran tema porque este es el barrio donde vive Bill Gates e imaginemos que no hay mucha poblacion viviendo. es lógico que el sueldo de Bill sea mucho más grande por lo que va a tirar hacia arriba la medición.

En este caso, el salario de Bill, es un valor atípico ya que no es frecuente dentro de nuestras mediciones.

Fijense que los valores atípicos dependen del contexto y del negocio que se esté analizando porque en este caso, no es malo para mi muestra que exista esta persona en mi dataset.

(Ejemplo del sensor de temperatura)

Existen varias formas de detectar estos valores atípicos, por lo general se trata de detecciones experimentales de las que hablaremos más adelante.

- Rango intercuartil
- Empirical Cumulative distribution-based outlier detection

```python
np.random.seed(0)
heights = np.random.normal(loc=170, scale=10, size=12) 
heights_with_outliers = np.concatenate([heights, [220, 220, 210, 215]])

sns.histplot(heights_with_outliers, bins=30, color='skyblue')

mean_height = heights_with_outliers.mean()

plt.axvline(mean_height, color='red', linestyle='--', linewidth=2, label='Media')
plt.xlabel('Altura (cm)')
plt.ylabel('Frecuencia')
plt.title('Distribución de Alturas con Outliers')

plt.legend()

# Mostrar el gráfico
plt.show()
print(mean_height)
```

### Media ponderada

Es una medida de tendencia central que es apropiada cuando en un conjunto de datos cada uno tiene una importancia relativa respecto a los demás datos. Se obtiene multiplicando cada uno de los datos por su ponderación y luego se suman

$\bar{x} = \frac{\Sigma_{i=1}^n x_i * w_i}{\Sigma_{w_i}}$

Un ejemplo típico de esto es calcular la nota final de un curso donde cada examen tiene un peso relativo a esta nota.

### Mediana

Es el valor que divide un conjunto de datos **ordenado** en 2 partes iguales.

¿Como sacarla?

- Ordenadar los datos de menor a mayor
- Calcular el valor central de ese conjunto

**Ejemplo de edades:**

Edades =  {22, 30, 29, 41, 35}

Edades_Ordenada =  {22, 29, 30, 35, 41}

Mediana = 30

**Ejemplo de tickets de compra**

Tickets = {120, 150, 80, 200, 170}

Tickets_ordenados = { 80, 120, 150, 170, 200 }

Mediana = 150

Tickets_ordenados_2 = { 80, 120, 150, 170, 200 , 380}

Mediana = 160

### Ventajas de usar la mediana:

- Robustez frente a outliers: No se ve afectada por valores extremos. Es ideal para usar cuando nuestro conjunto tiene outliers o **distribuciones sesgadas como ingresos de una población, tickets de compra, etc.**

- Representación de datos sesgados: En distribuciones de cola larga proporciona una mejor representación del centro de los datos (útil en distribuciones no simétricas: Duración de llamadas de un call-center, tasa de aparición de publicidad en una página web, etc)

**Caso de salario con outlier: Media y mediana**

Ingreso en Miles de dolares = { 30, 35, 40, 45, 1000  }

**Media:** (30 + 35 + 40 + 45 + 1000) / 5 = 230

**Mediana:** 40

### Moda

La moda es el valor que aparece con mayor frecuencia en un conjunto de datos. Un conjunto puede ser unimodal, bimodal o polimodal. 

Es especialmente útil en

- Datos categóricos: Es la única medida de tendencia central que se puede usar para datos categóricos nominales. Preferencia de colores entre un grupo de personas, categoría de compra preferida, Si tenemos usuarios activos segmentados por categorías (compra a la mañana, tarde o noche)
- Datos discretos: Número de veces que ocurre un evento

## Calculo de la moda

```python
data = {
    'color_preferido': ['Rojo', 'Azul', 'Verde', 'Rojo', 'Rojo', 'Azul', 'Amarillo', 'Verde', 'Azul', 'Rojo']
}

df = pd.DataFrame(data)

# Calcular la moda
moda = df['color_preferido'].mode()
print(f"Moda: {moda}")

# Contar las frecuencias de cada categoría
frecuencias = df['color_preferido'].value_counts()

# Crear el gráfico
plt.figure(figsize=(10, 6))
sns.barplot(x=frecuencias.index, y=frecuencias.values, palette='viridis')

# Añadir detalles al gráfico
plt.title('Frecuencia de Preferencias de Colores')
plt.xlabel('Color')
plt.ylabel('Número de Personas')
plt.xticks(rotation=45)
plt.show()
```

### Mediana y Media en distribucion con outliers

```python
salarios = [30, 35, 40, 45, 40, 30, 45, 1000]

# Calcular la media y la mediana
media = np.mean(salarios)
mediana = np.median(salarios)

# Crear un gráfico de dispersión con líneas para la media y la mediana
plt.figure(figsize=(8, 6))

# Histograma de los salarios
sns.histplot(salarios, bins=30, kde=False, color='skyblue')

# Añadir líneas para la media y la mediana
plt.axvline(media, color='red', linestyle='--', label=f'Media: {media:.2f}')
plt.axvline(mediana, color='green', linestyle='-', label=f'Mediana: {mediana:.2f}')

# Etiquetas y título
plt.xlabel('Ingreso (Miles de dólares)')
plt.ylabel('Frecuencia')
plt.title('Distribución de Ingresos con Outlier')

# Mostrar leyenda
plt.legend()

# Mostrar el gráfico
plt.show()

# Imprimir los resultados de la media y la mediana
print(f"Media: {media:.2f} miles de dólares")
print(f"Mediana: {mediana:.2f} miles de dólares")
```

# Variabilidad y dispersión

Las medidas de dispersión indican el grado de concentración de datos de una V.A al rededor de una medida de posición

Las medidas más utilizadas

- **Rango**: Es un valor numérico que mide la diferencia entre el valor mas alto y el más bajo en un conjunto de datos mostrando cuanto varian esos valores.
- **Varianza:** Es una medida estadística que describe la dispersión de un conjunto de datos en relación a su media. Nos dice que tan alejados estan los valores indivduales respecto a la media

Var(X) = $\frac{\Sigma_{i = 1} (x_i - \mu)^2}{n}$

$x_i$ es cada valor individual del conjunto de datos

Una varianza baja indica que los valores están cerca de su media (poca dispersión)

La varianza tiene alguna dificultad en cuanto a su interpretación ya que en su formula tiene un término cuadrado lo que eleva la unidad al cuadrado.

- **Desvío estándar**: Es la raíz cuadrada de la varianza. Lo que da un número más pequeño y más sencillo de interpretar a efectos prácticos


> La varianza por su naturaleza de cálculo es sensible a outliers. Además, como depende de la escala, es dificil de comparar. Para esto existen otras medidas como CV (Coeficiente de variación)

$CV = \sigma / \mu $

En general se usa el desvío estándar para interpretar los datos ya que las unidades están en concordancia con los datos

```python
np.random.seed(0)
low_variance = np.random.normal(loc=50, scale=2, size=100)

# Generar datos con mucha varianza
high_variance = np.random.normal(loc=50, scale=10, size=100)

# Crear el gráfico
plt.figure(figsize=(10, 6))

# Gráfico de poca varianza con curva KDE
sns.kdeplot(low_variance, label='Poca Varianza', color='blue', fill=True, alpha=0.5)

# Gráfico de mucha varianza con curva KDE
sns.kdeplot(high_variance, label='Mucha Varianza', color='red', fill=True, alpha=0.5)
plt.axvline(low_variance.mean(), color='blue', linestyle='--')
# Etiquetas y leyenda
plt.title('Comparación de Varianza')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()

# Mostrar el gráfico
plt.show()
print("Varianza de low_variance" ,np.var(low_variance))
print("Varianza de high_variance" ,np.var(high_variance))
print("Desvío estándar de low_variance" ,np.std(low_variance))
print("Desvio de high_variance" ,np.std(high_variance))
```

# Percentiles
El percentil es otra medida de dispersión aunque la mediana es una medida de tendencia central, los percentiles se usan para describir como se distribuyen los datos a lo largo del rago y para identificar la dispersión.

Es una medida que divide un conjunto de datos ordenados en 100 partes iguales. El percentil indica el valor por debajo del cual se encuentra un determinado porcentaje de datos

```python
np.random.seed(40)
mean = 170
sigma = 10
n = 1000

heights = np.random.normal(loc=mean, scale = sigma, size = n)

q1 = np.percentile(heights, 25)
q2 = np.percentile(heights, 50)
q3 = np.percentile(heights, 75)
p90 = np.percentile(heights, 90)

plt.figure(figsize=(10,6))
sns.histplot(heights, bins=30, kde=True, color='red')

plt.axvline(q1, color='blue', linestyle='--', label=f'Q1 (25%) = {q1:.2f} cm')
plt.axvline(q2, color='green', linestyle='--', label=f'Q2 (Mediana) = {q2:.2f} cm')
plt.axvline(q3, color='red', linestyle='--', label=f'Q3 (75%) = {q3:.2f} cm')
plt.axvline(p90, color='purple', linestyle='--', label=f'P90 (90%) = {p90:.2f} cm')


# Añadir etiquetas y título
plt.title('Distribución Normal de Alturas', fontsize=16)
plt.xlabel('Altura (cm)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.legend()

# Mostrar el gráfico
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# Configuración inicial
np.random.seed(40)
alpha = 2.0  # Parámetro de forma (shape)
beta = 50.0  # Parámetro de escala (scale)
n = 1000     # Número de consumidores


gastos = np.random.gamma(shape=alpha, scale=beta, size=n)


q1 = np.percentile(gastos, 25)
q2 = np.percentile(gastos, 50)
q3 = np.percentile(gastos, 75)
p90 = np.percentile(gastos, 90)

plt.figure(figsize=(10, 6))
sns.histplot(gastos, bins=30, kde=True, color='orange')

plt.axvline(q1, color='blue', linestyle='--', label=f'Q1 (25%) = ${q1:.2f}', linewidth=2.5)
plt.axvline(q2, color='green', linestyle='--', label=f'Q2 (Mediana) = ${q2:.2f}', linewidth=2.5)
plt.axvline(q3, color='red', linestyle='--', label=f'Q3 (75%) = ${q3:.2f}', linewidth=2.5)
plt.axvline(p90, color='purple', linestyle='--', label=f'P90 (90%) = ${p90:.2f}', linewidth=2.5)


plt.title('Distribución Gamma de Gastos de Consumidores', fontsize=16)
plt.xlabel('Gastos ($)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.legend()


plt.tight_layout()
plt.show()
```

# Diagramas de caja

Es un método estandarizado para representar gráficamente datos numéricos a través de sus cuartiles (q1, q2, q3) La idea es visualiar la mediana y los cuartiles y también sus valores atípicos

Para crear un diagrama de caja:
- Rango (sin datos atípicos)
- Datos atípicos
- Rango intercuartil (IQR)
- Cuartiles (q1,q2,q3)
- Mediana
- Mínimo y máximo

```python
sns.boxplot(heights)
```

```python
np.random.seed(40)

# Definir medias y desviaciones estándar para diferentes países
countries = {
    'Argentina': {'mean': 170, 'sigma': 10},
    'Colombia': {'mean': 165, 'sigma': 8},
    'Uruguay': {'mean': 175, 'sigma': 9},
    'Brasil': {'mean': 172, 'sigma': 11},
    'México': {'mean': 168, 'sigma': 10}
}

# Número de personas por país
n = 1000

# Generar datos de alturas para cada país
data = []
for country, params in countries.items():
    heights = np.random.normal(loc=params['mean'], scale=params['sigma'], size=n)
    country_data = pd.DataFrame({'País': country, 'Altura (cm)': heights})
    data.append(country_data)

# Concatenar los datos de todos los países en un solo DataFrame
df_heights = pd.concat(data)

# Crear diagramas de caja para comparar las alturas por país
plt.figure(figsize=(12, 8))
sns.boxplot(x='País', y='Altura (cm)', data=df_heights, palette='Set2')

# Añadir título y etiquetas
plt.title('Comparación de Alturas entre Países', fontsize=16)
plt.xlabel('País', fontsize=12)
plt.ylabel('Altura (cm)', fontsize=12)

# Mostrar el gráfico
plt.show()
```

## Diagramas de violin

Se utiliza para visualizar la distribución de los datos y su densidad de probabilidad. Es una combinación entre un diagrama de cajas y un histograma girado. La barra en el centro representa el intervalo **intercuartil** y la negra fina que se extiende representa el 95% de los intervalos de confianza.

El punto negro es la mediana.

Este diagrama es útil para ver una representación más acertada de los datos ya que el diagrama de caja y bigotes, por su simplicidad, tiende a ocultar detalles significativos.

```python
np.random.seed(40)

# Definir medias y desviaciones estándar para diferentes países
countries = {
    'Argentina': {'mean': 170, 'sigma': 10},
    'Colombia': {'mean': 165, 'sigma': 8},
    'Uruguay': {'mean': 175, 'sigma': 9},
    'Brasil': {'mean': 172, 'sigma': 11},
    'México': {'mean': 168, 'sigma': 10}
}

# Número de personas por país
n = 1000

# Generar datos de alturas para cada país
data = []
for country, params in countries.items():
    heights = np.random.normal(loc=params['mean'], scale=params['sigma'], size=n)
    country_data = pd.DataFrame({'País': country, 'Altura (cm)': heights})
    data.append(country_data)

# Concatenar los datos de todos los países en un solo DataFrame
df_heights = pd.concat(data)

# Crear diagramas de caja para comparar las alturas por país
plt.figure(figsize=(12, 8))
sns.violinplot(x='País', y='Altura (cm)', data=df_heights, palette='Set2')

# Añadir título y etiquetas
plt.title('Comparación de Alturas entre Países', fontsize=16)
plt.xlabel('País', fontsize=12)
plt.ylabel('Altura (cm)', fontsize=12)

# Mostrar el gráfico
plt.show()
```

## Rango Intercuartil IQR

En palabras simples es la diferencia entre el tercer y el primer cuartil de una distribución.

**IQR = Q3 - Q1**

A diferencia del rango o la varianza, por ejemplo, es estadístico robusto, es decir, no es sensible a los valores atípicos.

El IQR se usa para identificar valores atípicos en una sola muestra. Se consideran outliers aquellos puntos de datos que están por debajo de Q1 - 1.5IQR o por encima de Q3 + 1.5IQR
